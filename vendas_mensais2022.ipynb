```python
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(vendas_2022)
df.index = lojas
df
```

```python
import matplotlib.pyplot as plt
```

```python

cores = ['yellow', 'blue', 'green', 'purple']

fig , axs = plt.subplots(2,2, figsize=(15,10))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas mensais das lojas A, B, C e D em 2022')

axs[0,0].plot(df.loc['A'], color='r', lw=2)
axs[0,0].set_title('Vendas de A', fontsize= 14, loc='left')

axs[0,1].plot(df.loc['B'], color='g', lw=2)
axs[0,1].set_title('Vendas de B',fontsize= 14, loc='left')

axs[1,0].plot(df.loc['C'], color='b', lw=2)
axs[1,0].set_title('Vendas de C',fontsize= 14, loc='left')

axs[1,1].plot(df.loc['D'], color='y', lw=2)
axs[1,1].set_title('Vendas de D', fontsize= 14, loc='left')

ymin = 0
ymax = 450

for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)
    ax.set_xlabel('MÃªs', fontsize = 16)
    ax.set_ylabel('Vendas',fontsize = 16)
    ax.grid(True)

plt.show()



```
